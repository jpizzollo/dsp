**Exercise:** The numbers generated by `numpy.random.random` are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from `numpy.random.random` and plot their PMF.  What goes wrong?

Now plot the CDF. Is the distribution uniform?

**Answer Ex4.2:** The PMF plot doesn't do a good job of showing the shape of the distribution because, rather than connecting points with non-zero probability, the step function draws a vertical line down to zero on the y-axis (zero probability) for values that are not part of the sample. Thus, in the sample of 1000 random numbers there are 2000 vertical lines that occlude the shape of the distribution. The CDF does a better job, showing a straight line, indicating that values between 0 and 1 are generated at uniform frequency.   

```python
from __future__ import print_function, division
%matplotlib inline
import numpy as np
import nsfg
import first
import thinkstats2
import thinkplot

random_numbers = np.random.random(1000)

random_pmf = thinkstats2.Pmf(random_numbers, label='random_pmf')
thinkplot.PrePlot(1, cols=2)
thinkplot.Pmf(random_pmf)
thinkplot.Config(xlabel='random number', ylabel='Probability')

random_cdf = thinkstats2.Cdf(random_numbers, label='random_cdf')
thinkplot.PrePlot(1)
thinkplot.SubPlot(2)
thinkplot.Cdf(random_cdf)
thinkplot.Config(xlabel='random number', ylabel='CDF')
```
